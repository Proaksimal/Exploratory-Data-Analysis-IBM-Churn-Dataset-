### Histogram + KDE plot for knowing the distribution form

```python
# adjust the figure size for better readability
plt.figure(figsize=(10,6))

features = numericals
for i in range(0, len(features)):
    plt.subplot(2, len(features)//2 + 1, i+1)
    sns.histplot(x=telco_churn[features[i]], color='skyblue', kde=True)
    plt.xlabel(features[i])
    plt.tight_layout()
```

Observations:


1.   SeniorCitizen column is indeed boolean/binary column, so we don't need to interpret the symmetrical.
2.   Tenure column is approximately symmetric, with two peaks or modes, indicating a bimodal distribution.
3. MonthlyCharges column is unexpectedly degree of symmetric, with a slight skew to the left, with 1 peak or mode, indicating a unimodal distribution.
4. TotalCharges is a skew to the right or positive skew, with 1 peak or mode on around 0 value. This indicates that majority of the customers on this dataset have relatively small total charges. This implies that most customers are likely new, as they haven't accrued significant charges yet.

## Question 1:

Who are the top 10 customers (based on customerID) that provide the highest total revenue to the company (based on TotalCharges) and still have ongoing subscription?

### Answer

```python
top10_customers = (telco_churn[telco_churn['Churn'] == 'No']
                .groupby('customerID')
                .agg(
                  sum_totalcharge = ('TotalCharges', 'sum')
                  )
                .reset_index()
                .sort_values('sum_totalcharge', ascending = False)
                .head(10))

top10_customers
```

```python
sns.set(style="whitegrid")
sns.set_palette("colorblind")

plt.figure(figsize=(10, 6))
top10_customers['customerID'] = top10_customers['customerID'].astype(str)
sns.barplot(data=top10_customers, x='customerID', y='sum_totalcharge')
plt.xticks(rotation=90)
```

```python
telco_churn[telco_churn['customerID'].isin(top10_customers['customerID'])]
```

## Question 2:

What is the average monthly charge for customers who have partner compared to those who don't?

### Answer:

```python
avg_monthly_charge = (telco_churn
                .groupby('Partner')
                .agg(
                  avg_monthly_charge = ('MonthlyCharges', 'mean')
                  )
                .reset_index())

avg_monthly_charge
```

```python
sns.set(style="whitegrid")
sns.set_palette("bright")

plt.figure(figsize=(10, 3))
sns.barplot(data=avg_monthly_charge, x='avg_monthly_charge', y='Partner', orient='h')
plt.xlabel('Average Monthly Charge')
plt.ylabel('Partner')
plt.title('Average Monthly Charge by Partner Status')
plt.show()
```

Observations:

The average monthly charge is higher for customers who have partner (67.85 USD) compared to customers who don't have partner (61.96 USD). Customers who have partner tend to pay higher monthly charges compared to customers who don't have partner. This may be because several factors, such as customers with partners being more likely to use more services or more expensive packages because of their needs.

## Question 3:

Based on customer gender, which gender is the most spender overall (Based on TotalCharges)?

### Answer:

```python
gender_total_charge = (telco_churn
                .groupby(['gender'])
                .agg(
                  sum_total_charge = ('TotalCharges', 'sum')
                  )
                .reset_index())

gender_total_charge
```

```python
labels = gender_total_charge['gender']
sizes = gender_total_charge['sum_total_charge']

plt.figure(figsize=(10, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140) # Note: autopct='%1.1f%% helps to add percentage label inside pie chart.
plt.title('Total Charges by Gender')
plt.axis('equal')
plt.show()
```

Observations:

It appears that males have spent a slightly higher amount overall compared to females.


*   Male customers have spent approximately 8,103,814.5 USD or approximately 50.5% of the total charges.
*   Female customers have spent approximately 7,952,354.2 USD or approximately 49.5% of the total charges.

Based on total charges, male customers are the higher spenders than female customers, slightly more than the total charges of female customers.


## Question 4:

How many customers have phone service enabled, and how does it relate to their churn status?

### Answer:

```python
total_phone_service = telco_churn[telco_churn['PhoneService'] == 'Yes'].shape[0]
total_phone_service_churn = telco_churn[(telco_churn['PhoneService'] == 'Yes') & (telco_churn['Churn'] == 'Yes')].shape[0]
total_phone_service_not_churn = telco_churn[(telco_churn['PhoneService'] == 'Yes') & (telco_churn['Churn'] == 'No')].shape[0]

print(f'Total customer with phone service (churn & not churn) = {total_phone_service}')
print(f'Total customer with phone service and churn = {total_phone_service_churn}')
print(f'Total customer with phone service and not churn = {total_phone_service_not_churn}')
```

Observations:

1. There are total of: 6352 customers that have phone service enabled.
2. Although the number of customers with phone service who churned is quite significant (1699 customers), there is still a much larger number of customers with phone service who remained subscribed (4653 customers). This indicates that having active phone service does not directly influence customers' decisions to churn. Other factors may influence customer churn decisions.
